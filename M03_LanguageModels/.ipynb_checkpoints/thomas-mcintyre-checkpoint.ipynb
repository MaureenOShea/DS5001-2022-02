```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.core.display import HTML
# This prevents matplotlib from throw irrelevant errors
from matplotlib.axes._axes import _log as matplotlib_axes_logger
matplotlib_axes_logger.setLevel('ERROR')
sns.set()
```

```python
OHCO = ['chap_num', 'para_num', 'sent_num', 'token_num']
```

```python
LINES = pd.DataFrame(open("../data/gutenberg/pg42324.txt", 'r', encoding='utf-8-sig').readlines(), columns=['line_str'])
LINES.index.name = 'line_num'
LINES.line_str = LINES.line_str.str.replace(r'\n+', ' ', regex=True).str.strip()
LINES.sample(10)
```

```python
title = LINES.loc[0].line_str.replace('The Project Gutenberg EBook of ', '')
print(title)
```

```python
## Clipping as demoed in class

clip_pats = [
    r"PREFACE.", 
    r"THE END."
]

pat_a = LINES.line_str.str.match(clip_pats[0])
pat_b = LINES.line_str.str.match(clip_pats[1])

line_a = LINES.loc[pat_a].index[0] - 2
line_b = LINES.loc[pat_b].index[0] - 1

print(line_a, line_b)

LINES = LINES.loc[line_a : line_b]

print(LINES.head(5))
print(LINES.tail(5))
```

```python
chap_pat = r"((?:letter|chapter)\s+\D+)|PREFACE."
chap_lines = LINES.line_str.str.match(chap_pat, case=False)
LINES.loc[chap_lines]
```

```python
LINES.loc[chap_lines, 'chap_num'] = [i+1 for i in range(LINES.loc[chap_lines].shape[0])]
LINES.loc[chap_lines]
```

```python
LINES.chap_num = LINES.chap_num.ffill()
LINES.sample(20)
```

```python
LINES = LINES.dropna(subset=['chap_num'])
LINES = LINES.loc[~chap_lines] 
LINES.chap_num = LINES.chap_num.astype('int') 
```

```python
LINES
```

```python
CHAPS = LINES.groupby(OHCO[:1]).line_str.apply(lambda x: '\n'.join(x)).to_frame('chap_str')
CHAPS['chap_str'] = CHAPS.chap_str.str.strip()
CHAPS.head(10)
```

```python
para_pat = r'\n\n+'
PARAS = CHAPS['chap_str'].str.split(para_pat, expand=True).stack()\
    .to_frame('para_str').sort_index()
PARAS.index.names = OHCO[:2]
PARAS['para_str'] = PARAS['para_str'].str.replace(r'\n', ' ', regex=True)
PARAS['para_str'] = PARAS['para_str'].str.strip()
PARAS = PARAS[~PARAS['para_str'].str.match(r'^\s*$')]
```

```python
PARAS.head()
```

```python
sent_pat = r'[.?!;:]+'
SENTS = PARAS['para_str'].str.split(sent_pat, expand=True).stack()\
    .to_frame('sent_str')
SENTS.index.names = OHCO[:3]
SENTS = SENTS[~SENTS['sent_str'].str.match(r'^\s*$')] # Remove empty paragraphs
SENTS.sent_str = SENTS.sent_str.str.strip()
```

```python
SENTS.head()
```

```python
token_pat = r"[\s',-]+"
TOKENS = SENTS['sent_str'].str.split(token_pat, expand=True).stack()\
    .to_frame('token_str')
TOKENS.index.names = OHCO[:4]
```

```python
TOKENS
```

```python
TOKENS['term_str'] = TOKENS.token_str.replace(r'[\W_]+', '', regex=True).str.lower()
VOCAB = TOKENS.term_str.value_counts().to_frame('n').reset_index().rename(columns={'index':'term_str'})
VOCAB.index.name = 'term_id'
```

```python
VOCAB
```

```python
vocab_list = VOCAB.term_str.to_list()
```

```python
# vocab_list
```

```python
def gather(ohco_level):
    global TOKENS_ALL
    level_name = OHCO[ohco_level-1].split('_')[0]
    df = TOKENS.groupby(OHCO[:ohco_level])\
        .token_str.apply(lambda x: x.str.cat(sep=' '))\
        .to_frame(f"{level_name}_str")
    return df
```

```python
sentence_list=gather(3).sent_str.to_list()
```
