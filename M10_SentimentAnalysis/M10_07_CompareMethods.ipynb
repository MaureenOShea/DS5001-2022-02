## Method C: DTM * SALEX

Use DTM. 

Turns out to be the same as B. Both B and C are different than A.

### C1: Absolute Mean &check;

````python
%%time
C1_chap = pd.DataFrame(index=DOC.index, columns=SALEX.columns)
for col in SALEX.columns:
    # C1_chap[col] = (DTM * SALEX2[col]).fillna(0).T.mean() # INCORRECT ... Wrong denominator
    C1_chap[col] = (DTM * SALEX2[col]).T.sum() / DOC.n # CORRECT
````

### C2: Relative Mean &cross;&cross;

````python
%%time
C2_chap = pd.DataFrame(index=DOC.index, columns=SALEX.columns)
for col in SALEX.columns:
    C2_chap[col] = (DTM * SALEX2[col]).mean(1)
````

## Method D: DTMTF * SALEX &check; 

Gallagher, et al. 

````python
BOW['tf'] = BOW.groupby(bag).n.apply(lambda x: x / x.sum())
````

````python
BOW.tf.head()
````

````python
%time DTMTF = BOW.tf.unstack()
````

````python
%%time
D_chap = pd.DataFrame(index=DOC.index, columns=SALEX.columns)
for col in SALEX.columns:
    D_chap[col] = (DTMTF * SALEX2[col]).sum(1)
````

## Compare by Plot

````python
# SALEX.columns
````

````python
emo_dfs = {}
for emo_col in SALEX.columns:
    comp_df = pd.concat([A1_chap[emo_col], 
                         A2_chap[emo_col],
                         B1_chap[emo_col], 
                         B2_chap[emo_col], 
                         C1_chap[emo_col], 
                         C2_chap[emo_col], 
                         D_chap[emo_col]], axis=1, 
                        keys=['a1', 'a2', 'b1','b2', 'c1', 'c2', 'd'])
    comp_df_norm = (comp_df - comp_df.mean()) / (comp_df.std())
    emo_dfs[emo_col] = comp_df_norm
COMPARE = pd.concat(emo_dfs.values(), keys=emo_dfs.keys(), axis=1)
````

## Correlate Methods

````python
def show_method_correlations(sent_col = 'nrc_polarity'):
    global COMPARE
    cols = list(COMPARE[sent_col].columns)
    pairs = pd.MultiIndex.from_product([cols, cols])
    fig, axes = plt.subplots(nrows=len(cols), ncols=len(cols), figsize=(25,25), sharex=False, clear=True)
    for pair in sorted(pairs):
        df = COMPARE[sent_col]
        x, y = pair
        row = cols.index(x)
        col = cols.index(y)
        if x == y:
            df.hist(x, ax=axes[row,col])
        else:
            corr = df[x].corr(df[y])
            df.plot.scatter(y, x, ax=axes[row,col], title=f"({y},{x})\ncorr={round(corr,2)}")
            axes[row,col].axis('off')
    fig.tight_layout()                    
````

## Create general function

````python
def compute_sentiment(ohco_level=1):
    global COMBO, SALEX, OHCO
    # OHCO = COMBO.index.names
    bag = OHCO[:ohco_level]
    DOC = COMBO.groupby(bag).term_str.count().to_frame('n')
    DOC_SENTS = (COMBO.groupby(bag)[SALEX.columns].sum().T / DOC.n).T
    # DOC_SENTS = COMBO.groupby(bag)[SALEX.columns].mean(1) # Old way
    return DOC_SENTS
````

## Viz functions

````python
def show_heatmap(X, height=600):
    fig = px.imshow(X, color_continuous_scale='Spectral', height=height)
    fig.update_layout(coloraxis_showscale=False)
    fig.update_xaxes(side="top", tickangle=45)
    fig.show()
````

````python
def show_rolling(df, title='', sent_col='polarity', w=5, cuts=25, ax=None, figsize=(20, 20)):
    df = df.reset_index(drop=True).reset_index().rename(columns=dict(index='seq'))
    df['cut'] = pd.cut(df.seq, cuts)
    df = df.groupby('cut')[sent_col].mean().reset_index(drop=True)
    plt = df.fillna(0).rolling(w).mean().plot(figsize=figsize, legend=True, fontsize=14, rot=45, title=title, ax=ax)
````
