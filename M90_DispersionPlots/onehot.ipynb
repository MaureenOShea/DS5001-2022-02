```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
%matplotlib inline
```

```python
OHCO = "chap_num para_num sent_num token_num".split()
```

```python
tokens = pd.read_csv('../MOD02--TextModels/austen-persuasion.csv').set_index(OHCO)
```

```python
tokens
```

```python
tokens['term_str'] = tokens.token_str.str.replace(r'\W+', '').str.lower()
```

```python
onehot = pd.get_dummies(tokens, columns=['term_str'], prefix='', prefix_sep='', drop_first=True).reset_index(drop=True).iloc[:,1:]
```

```python
onehot['anne'].plot(figsize=(20, 1))
```

```python
onehot[['love','hate', 'joy', 'happiness']].plot(figsize=(20, .5))
```

```python
X = onehot.reset_index()

# set size of figure
plt.figure(figsize=(22,6))

# use horizontal stripplot with x marker size of 5
sns.stripplot(y='yours', x='index', data=X,
 orient='h', marker='X', color='navy', size=5)

# rotate x tick labels
plt.xticks(rotation=15)

# remover borders of plot
plt.tight_layout()
plt.show()
```
